```python
# --- Core Libraries ---
import pandas as pd
import numpy as np

# --- Visualization ---
import matplotlib.pyplot as plt
import seaborn as sns

# --- Settings ---
# Establecer un estilo visual para los gráficos
sns.set_style('whitegrid')

# --- Data Loading ---
# Definir el path relativo al archivo de datos
# Asumiendo que el notebook está en 'notebooks/' y los datos en 'data/'
file_path = '../data/concrete_data.csv'
df_concrete = pd.read_csv(file_path)
```

```python
df_concrete.head(5)
```

```python
print(" INFORMACION DEl DATAFRAME ")
df_concrete.info()
```

```python
print(" CHEQUEO DE VALORES NULOS ")
print(df_concrete.isnull().sum())
```

```python
print("Descriptive Statistics:")
print(df_concrete.describe())
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(df_concrete['concrete_compressive_strength'], kde=True, color='skyblue', bins=30)
plt.title('Distribution of Concrete Strength')
plt.xlabel('Strength MPa')
plt.ylabel('Frequency')
plt.show()
```

```python
corr_matrix = df_concrete.corr()

plt.figure(figsize=(12, 8))
sns.heatmap(corr_matrix, annot=True, fmt=".2f", cmap='coolwarm', linewidths=0.5)
plt.title('Matriz de correlacion de caracteristicas')
plt.show()
```

```python
print("Correlacion con Sthength:")
print(corr_matrix['concrete_compressive_strength'].sort_values(ascending=False))
```
